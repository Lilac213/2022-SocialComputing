```python
import pandas as pd
import os
def read_Data(data_path):
    data_list = []

    for file in os.listdir(data_path):
        data_list.append(pd.read_csv(os.path.join(data_path, file), index_col=False))

    data = pd.concat(data_list)
    return data
```

```python
data = read_Data('create_data.csv')
len(data)
```

```python
data.to_csv('create_data.csv', index=False)
```
