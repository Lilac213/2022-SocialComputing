### 根据活跃用户找合作的人

```python
res_users = pd.read_csv('./res_users_1k.csv')
```

```python
users = res_users[["actor_id"]].values.reshape(1000).tolist()
# print(users)
```

```python
antusers = {}
i = 0
for user in users:
    i += 1
    antusers[user] = i
# print(antusers)
```

```python
mat = [[0 for k in range(1001)]]
```

```python
for i in range(1000):
    user = users[i]
    if i==1:
        mat.append([0 for k in range(1001)])
        continue
    s = str(user)
    print(time.time())
    result = o.execute_sql("select first.actor_id, count(*) from ods_github_log first, ods_github_log second, temp_active_users_2021_1k where first.actor_id=temp_active_users_2021_1k.actor_id and first.pt>='20210101' and first.pt<'20220101' and second.pt>='20210101' and second.pt<'20220101' and first.repo_id=second.repo_id and first.actor_id<>second.actor_id and second.actor_id=" + s + " group by first.actor_id order by first.actor_id;",hints={'odps.sql.validate.orderby.limit': 'false'})
    links = result.open_reader().to_pandas().values.tolist()
#     print(links)
    row = [0 for k in range(1001)]
    for j in links:
        row[antusers[j[0]]] = j[1]
    mat.append(row)
#     print(mat)
#     print(time.time())
    if i % 10 == 0:
        matnp = pd.DataFrame(data=mat)
        matnp.to_csv("./mat.csv")
        print(i)
#     break
```
